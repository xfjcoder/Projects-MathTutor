```python
df = pd.read_pickle("problems_with_embeddings.pkl")
```

```python
# Store in FAISS for similarity search
dimension = len(df["problem_embedding"][0])
index = faiss.IndexFlatL2(dimension)
faiss_data = np.array(df["problem_embedding"].tolist(), dtype=np.float32)
index.add(faiss_data)
```

```python
# Store solutions in SQLite for structured retrieval
conn = sqlite3.connect("problems.db")
df.to_sql("problems", conn, if_exists="replace", index=False)
conn.close()
```

```python
# Function to find similar problems
model = SentenceTransformer("all-MiniLM-L6-v2")
def retrieve_similar_problems(query, top_k=3):
    query_embedding = model.encode(query).reshape(1, -1).astype(np.float32)
    distances, indices = index.search(query_embedding, top_k)
    similar_problems = df.iloc[indices[0]][["problem_id", "problem"]]
    return similar_problems
```

```python
query = "Fifteen integers $a_1, a_2, a_3, \dots, a_{15}$ are arranged in order on a number line. The integers are equally spaced and have the property that\[1 \le a_1 \le 10, \thickspace 13 \le a_2 \le 20, \thickspace \text{ and } \thickspace 241 \le a_{15}\le 250.\]What is the sum of digits of $a_{14}?$?"
similar_problems = retrieve_similar_problems(query)
similar_problems
```

```python
# Function to get solutions for a specific problem_id
def get_solutions(problem_id):
    query = """
    SELECT solution, letter, answer
    FROM problems
    WHERE problem_id = ?
    """
    conn = sqlite3.connect("problems.db")
    solutions = pd.read_sql(query, conn, params=(problem_id,))
    conn.close()  # ✅ Always close the connection
    return solutions
```

```python
# Function to generate explanation using LLM
from openai import OpenAI
client = OpenAI()

def explain_solution(problem, solution):
    prompt = f"""
    The problem has a solution. Explain{solution} step by step in 250 words or less.

    Problem: {problem}

    Solution: {solution}

    Explanation:
    """
    response = client.chat.completions.create(
        model="gpt-4",
        messages=[{"role": "system", "content": "You are an expert math tutor."},
                  {"role": "user", "content": prompt}]
    )
    return response.choices[0].message
```

```python
# Function to provide hints
def generate_hint(problem, solution):
    prompt = f"""
    Give a hint for solving the following problem without revealing the full {solution}:

    Problem: {problem}

    Hint:
    """
    response = client.chat.completions.create(
        model="gpt-4",
        messages=[{"role": "system", "content": "You are a math tutor providing hints."},
                  {"role": "user", "content": prompt}]
    )
    return response.choices[0].message
```

```python
# Function to evaluate user’s answer
def evaluate_answer(user_answer, correct_answer):
    prompt = f"""
    Evaluate the student's answer against the correct solution.

    Student's Answer: {user_answer}
    Correct Answer: {correct_answer}

    Feedback:
    """
    response = client.chat.completions.create(
        model="gpt-4",
        messages=[{"role": "system", "content": "You are a math tutor who provides feedback on student answers."},
                  {"role": "user", "content": prompt}]
    )
    return response.choices[0].message
```

```python
# Interactive Gradio chatbot function
def chat_interface(user_query, mode="explanation"):
    similar_problems = retrieve_similar_problems(user_query)
    response_text = ""

    for _, row in similar_problems.iterrows():
        response_text += f"\n🔹 **Problem:** {row['problem']}\n"
        solutions = get_solutions(row["problem_id"])

        if len(solutions) > 1:
            response_text += "\nThere are multiple solutions available. Would you like to see an alternative solution? (yes/no)\n"

        if mode == "evaluate":
            return "Please enter your solution, and I'll evaluate it against the correct answer."

        for i, sol in solutions.iterrows():
            response_text += f"\n💡 **Solution {i+1}:** {sol['solution']}\n"
            response_text += f"\n✅ **Answer ({sol['letter']}):** {sol['answer']}\n"

            if mode == "hint":
                hint = generate_hint(row["problem"], sol['solution'])
                return f"💡 **Hint:** {hint.content}"

            if mode == "explanation":
                explanation = explain_solution(row["problem"], sol["solution"])
                response_text += f"\n📖 **Step-by-Step Explanation:** {explanation.content}\n"
                response_text += "-" * 70 + "\n"
                #break  # Show only one solution initially

        return response_text or "No matching problems found."
```

```python
user_query = "On Monday, Taye has $$2$. Every day, he either gains $$3$ or doubles the amount of money he had on the previous day. How many different dollar amounts could Taye have on Thursday, $3$ days later?"
chat_response = chat_interface(user_query, mode="explanation")
pprint.pprint(chat_response)
```

```python
# Function for evaluating user's response
def evaluate_mode(user_input, problem_query):
    similar_problems = retrieve_similar_problems(problem_query)
    if similar_problems.empty:
        return "No matching problem found to evaluate."

    problem_id = similar_problems.iloc[0]["problem_id"]
    solutions = get_solutions(problem_id)
    correct_answer = solutions.iloc[0]["answer"] if not solutions.empty else "No correct answer found."

    feedback = evaluate_answer(user_input, correct_answer)
    return feedback.content
```
